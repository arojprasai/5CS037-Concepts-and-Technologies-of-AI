```python
if 'Age' in df.columns:
    df.loc[:, 'Age'] = df['Age'].fillna(df['Age'].mean())
if 'Height' in df.columns:
    df.loc[:, 'Height'] = df['Height'].fillna(df['Height'].mean())
df.head()
```

```python
if 'Weight' in df.columns:
    df.loc[:, 'Weight'] = df['Weight'].fillna(df['Weight'].mean())
if 'BMI' in df.columns:
    df.loc[:, 'BMI'] = df['BMI'].fillna(df['BMI'].mean())
if 'Temperature' in df.columns:
    df.loc[:, 'Temperature'] = df['Temperature'].fillna(df['Temperature'].mean())
df = df.rename(columns={'Heart Rate': 'Heart_Rate', 'Blood Pressure': 'Blood_Pressure', 'Blood Type':'Blood_Type'})
df.columns
```

```python
if 'Heart_Rate' in df.columns:
    df.loc[:, 'Heart_Rate'] = df['Heart_Rate'].fillna(df['Heart_Rate'].mean())
if 'Blood_Pressure' in df.columns:
    df.loc[:, 'Blood_Pressure'] = df['Blood_Pressure'].fillna(df['Blood_Pressure'].mean())
if 'Cholesterol' in df.columns:
    df.loc[:, 'Cholesterol'] = df['Cholesterol'].fillna(df['Cholesterol'].mean())
if 'Cholesterol' in df.columns:
    df.loc[:, 'Cholesterol'] = df['Cholesterol'].fillna(df['Cholesterol'].mean())
if 'Diabetes' in df.columns:
    df.loc[:, 'Diabetes'] = df['Diabetes'].fillna(df['Diabetes'].mode()[0])
if 'Smoking' in df.columns:
    df.loc[:, 'Smoking'] = df['Smoking'].fillna(df['Smoking'].mode()[0])
df.head()
```

```python
null_values_after_correction = df.isnull().sum()
print(null_values_after_correction)
```

```python
df = df.rename(columns={'Student ID': 'Student_ID'})
if 'Student_ID' in df.columns:
  df.loc[:, 'Student_ID'] = range(1, len(df) + 1)
df.head(10)
```

```python
null_values_after_correction = df.isnull().sum()
print(null_values_after_correction)
```

```python
df = df.dropna(subset=['Gender', 'Blood_Type'])
df.shape
```

```python
import matplotlib.pylab as plt
import seaborn as sns
df = pd.read_csv("/content/drive/MyDrive/Dataset/Titanic-Dataset.csv")
df.head()
```

```python
filtered_df = df[df['Pclass'] == 1]

selected_column = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
new_df = filtered_df[selected_column]
```

```python
print(f"Min: {new_df['Fare'].min()}")
print(f"Max: {new_df['Fare'].max()}")
print(f"Mean: {new_df['Fare'].mean()}")
print(f"Median: {new_df['Fare'].median()}")
```

```python
null_in_age = new_df.loc[:, 'Age'].isnull().sum()
print(null_in_age)
```

```python
new_df.dropna(subset=['Age'])
```

```python
null_value_age = new_df.isnull().sum()
print(null_value_age)
```

```python
new_df = new_df.dropna(subset=['Age'])
null_value_age = new_df.isnull().sum()
print(null_value_age)
```

```python
df['Embarked']
```

```python
df['Embarked'] = df['Embarked'].replace({
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
})
print(df['Embarked'])
```

```python
df.head()
```

```python
encoded_columns = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, encoded_columns], axis=1)
df = df.drop(columns=['Embarked'])
df.head()
```

```python
mean_survival_rates = df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_rates)
```

```python
#Bar plot for mean survival rates
plt.subplot(1,2,1)
sns.barplot(x=mean_survival_rates.index, y=mean_survival_rates.values)
plt.title('Mean Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Survival Rate')
#Box plot for survival distribution by gender
plt.subplot(1,2,2)
sns.boxplot(x='Sex', y='Survived', data=df)
plt.title('Survival Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.tight_layout()
plt.show()
```

```python
null_df = df.isnull().sum()
print(null_df)
```

```python
g = sns.catplot(
    data=df,
    x='Sex',
    y='Survived',
    hue='Embarked_Cherbourg',
    kind='bar',
    aspect=1,
    height=5
)
g.set_titles('{COL_NAME}')
g.set_axis_labels('Sex', 'Survived')
g.set_xticklabels(['Male', 'Female'])
g.fig.suptitle('Survival Distribution by Gender and Port of Embarkation', y=1.02)
plt.show()
```
